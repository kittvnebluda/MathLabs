```python
import numpy as np
from matplotlib import pyplot as plt

def f(x, c):
    return x**2/2-(x-c/2)**2
```

```python
x = np.linspace(-10, 10, 500)

for c in range(-5, 6):
    plt.plot(x, f(x, c), label=f"С={c}")

plt.plot(x, x**2/2, label="Особ. реш.", color="red")

# Добавление легенды
plt.legend(loc='upper right')

# Добавление заголовка и меток к осям
plt.title('Интегральные кривые')
plt.xlabel('x')
plt.ylabel('y')

# Отображение графиков
plt.grid()
plt.show()
```
