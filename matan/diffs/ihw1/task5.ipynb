```python
import numpy as np
from matplotlib import pyplot as plt

def f(x, c):
    return x**2/2-(x-c/2)**2
```

```python
x = np.linspace(-10, 10, 500)

for c in range(-5, 6):
    plt.plot(x, f(x, c), label=f"={c}")

plt.plot(x, x**2/2, label=". .", color="red")

#  
plt.legend(loc='upper right')

#      
plt.title(' ')
plt.xlabel('x')
plt.ylabel('y')

#  
plt.grid()
plt.show()
```
