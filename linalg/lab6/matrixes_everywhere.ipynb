```python
import numpy as np
from numpy import sin, cos, sqrt, pi
from matplotlib import pyplot as plt

def rot_z_mat(phi):
    return np.array([[cos(phi), -sin(phi)], [sin(phi), cos(phi)]])

a, b, c, d = 2, -7, 7, -6

x = np.linspace(-40, 40, 500)
```

Задание 1

```python
sim_y_mat = np.array([[1, 0], [0, -1]])
task_mat = rot_z_mat(np.arctan(a)) @ sim_y_mat @ rot_z_mat(-np.arctan(a))
print(f"Итоговая матрица:\n{task_mat}")

self_vecs = np.array([[0.5, 1], [-2, 1]])

# Координаты начала вектора
origin = np.array([0,0])

# Координаты векторов
vecs1 = [np.array([2, 1]), np.array([2, -1]), np.array([-2, 1]), np.array([-2, -1]),
         np.array([2, 0]), np.array([-2, 0]), np.array([0, 1]), np.array([0, -1])]
vecs2 = [task_mat @ vec for vec in vecs1]

# Создание объекта рисунка и осей
_, ax = plt.subplots()

# Нарисовать вектор с началом в origin
for i, vec in enumerate(vecs2, start=1):
    ax.quiver(*origin, *vec, angles='xy', scale_units='xy', scale=1, color=f"blue")

for i, vec in enumerate(vecs1, start=1):
    ax.quiver(*origin, *vec, angles='xy', scale_units='xy', scale=1, color=f"red")

ax.quiver(*origin, *self_vecs[0], angles='xy', scale_units='xy', scale=1, color=f"green", label=f"Собственные векторы")
ax.quiver(*origin, *self_vecs[1], angles='xy', scale_units='xy', scale=1, color=f"green")

# Настроить пределы осей
ax.set_xlim(-3, 3)
ax.set_ylim(-3, 3)

# Добавить метки к осям
ax.set_xlabel('X')
ax.set_ylabel('Y')

plt.plot(x, a*x, label=f"Прямая y={a}x")

# Показать график
plt.legend()
plt.grid()
plt.show()
```
