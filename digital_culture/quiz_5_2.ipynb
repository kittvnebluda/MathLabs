```python
import pandas as pd

df = pd.read_csv("quiz_5_2_2_data.csv", names=["TIME", "SERIES"])
df.head()
```

```python
import matplotlib.pyplot as plt

plt.plot(df.TIME, df.SERIES)
plt.xlabel("TIME")
plt.ylabel("SERIES")
plt.show()
```

SMA

```python
window = 7
k = int((window - 1)/2)

average_df = df.copy()
for i in range(df.SERIES.size):
    average_df.SERIES[i] = 1/(2*k + 1) * sum(df.SERIES.loc[i - k: i + k])
average_df = average_df[k:-k]
print(average_df)
```

```python
plt.plot(average_df.TIME, average_df.SERIES)
plt.plot(df.TIME, df.SERIES)
```

```python
df2 = pd.read_csv("quiz_5_2_3_data.csv", names=["TIME", "SERIES"])
df2
```

EMA

```python
alpha = 0.3
average_df2 = df2.copy()
average_df2.SERIES[0] = df2.SERIES[0]
for i in range(1, df2.SERIES.size):
    average_df2.SERIES[i] = alpha*df2.SERIES[i] + (1-alpha) * average_df2.SERIES[i-1]
average_df2
```

```python
plt.plot(average_df2.TIME, average_df2.SERIES)
plt.plot(df2.TIME, df2.SERIES)
```
